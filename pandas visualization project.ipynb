```python
df1=pd.read_csv('USA_cars_datasets.csv')
df1
```

```python
df1.isnull().sum()
```

```python
#infer_objects attempts to infer better data types for object columns. 
df1.infer_objects().dtypes
```

```python
#group the rows in df by the "brand" column, and then calculate the sum of the "price" column for each group.
df3=df1[['brand','price']].groupby('brand')['price'].sum()
df3
```

```python
# select only the rows where the "price" column is less than 1000.
df_price_range=df1[df1['price']<1000]
df_price_range
```

```python
#sort the DataFrame in ascending order of the "price" column
df4=df_price_range[['brand','price']].groupby('brand')['price'].sum().sort_values()
df4
```

```python
#summary of the mean profit percentage for each combination of product and country.
data_pt_profit=pd.pivot_table(df1,index=("year"),columns=("model"),values=["price"],aggfunc=(np.mean))
data_pt_profit.sort_values(by=("year"),ascending=True)
```

```python
styled_table = data_pt_profit.style.format('{:.2f}%').applymap(lambda x: 'background-color: red' if x < 100 else '')
styled_table
```

```python
#this shows all brands and their mileage
all_brand=pd.pivot_table(df1,index="brand",values="mileage",aggfunc=np.sum)
all_brand
```
